```python
import pandas as pd
```

```python
def rgb2hex(rgb):
    r = int(rgb.split(',')[0])
    g = int(rgb.split(',')[1])
    b = int(rgb.split(',')[2])
    
    return "#{:02x}{:02x}{:02x}".format(r,g,b)

print(rgb2hex('150, 21, 0'))
```
