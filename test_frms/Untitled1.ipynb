```python
import pandas as pd
```

```python
df = pd.read_csv('Libro4.csv')
df
```

```python
df['cod_col'] = df['codigo'].map(lambda x:x.split(' ')[1][7:14])
df
```

```python
def collor(row):
    color = row.loc['cod_col']
    return ['background-color: {}'.format(color) for r in row]
```

```python
res = df.style.apply(collor, axis=1)
```
