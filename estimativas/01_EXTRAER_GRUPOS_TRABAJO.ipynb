```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# Ruta original del archivo Excel
path_xlsx_cod_cos = r'C:\Documents\Ingenio Azucarero Guabira S.A\UTEA - SEMANAL - AVANCE COSECHA\2025\ESTIMATIVAS\_DATA_ENVIAR\COS_COS.xlsx'
```

```python
# Cargar DataFrame
df_cod_cos = pd.read_excel(path_xlsx_cod_cos)
```

```python
# Carpeta donde quieres guardar las im√°genes de los grupos
output_folder = r'C:\Documents\Ingenio Azucarero Guabira S.A\UTEA - SEMANAL - AVANCE COSECHA\2025\ESTIMATIVAS\_DATA_ENVIAR\tablas_cod_cos'
os.makedirs(output_folder, exist_ok=True)  # crea la carpeta si no existe
```

```python
# Guarda tablas por grupo como im√°genes usando matplotlib
for grupo, df_grupo in df_cod_cos.groupby('cod_cos'):
    df_export = df_grupo[['inst', 'cod_ca', 'nom_ca']].copy()

    # Renombrar columnas
    df_export.columns = ['Institucion', 'Codigo', 'Nombre Ca√±ero']

    fig, ax = plt.subplots(figsize=(5, len(df_export)*0.1 + 1))
    ax.axis('off')

    tabla = ax.table(cellText=df_export.values,
                     colLabels=df_export.columns,
                     cellLoc='center',
                     loc='center')

    tabla.auto_set_font_size(False)
    tabla.set_fontsize(10)
    tabla.auto_set_column_width(col=list(range(len(df_export.columns))))

    # Poner en negrita las cabeceras (primera fila)
    for col in range(len(df_export.columns)):
        tabla[(0, col)].set_text_props(weight='bold')

    # Alinear columna "Nombre Ca√±ero" a la izquierda
    col_index = df_export.columns.get_loc('Nombre Ca√±ero')
    for i in range(len(df_export)):
        tabla[(i+1, col_index)].set_text_props(ha='left')

    plt.tight_layout()

    archivo_salida = os.path.join(output_folder, f'{grupo}.png')
    plt.savefig(archivo_salida, dpi=300, bbox_inches='tight')
    plt.close()

    print(f'Grupo {grupo} guardado como imagen en {archivo_salida}')

print('¬°Proceso completado correctamente!')
```
