```python
PATH_XLSX_AVANCE_COSECHA = r'G:\Ingenio Azucarero Guabira S.A\UTEA - SEMANAL - AVANCE COSECHA\2025\AVANCE SEMANAL\AVANCE DE COSECHA V1.xlsx'
PATH_XLSX_CONTACTOS_GRUPOS = r'G:\Ingenio Azucarero Guabira S.A\UTEA - SEMANAL - AVANCE COSECHA\2025\DATA\CONTACTOS_GRUPO.xlsx'
```

```python
df_avance_por_variedad = pd.read_excel(PATH_XLSX_AVANCE_COSECHA, sheet_name='AVANCE POR VARIEDADES')
df_contactos_grupos = pd.read_excel(PATH_XLSX_CONTACTOS_GRUPOS)
```

```python
# coloca la fila 1 como cabecera
df_avance_por_variedad.columns = df_avance_por_variedad.iloc[1]
# descarta las primeras 2 filas, (fila en blanco y fila de nombres de caberas)
df_avance_por_variedad = df_avance_por_variedad.iloc[2:].reset_index(drop=True)
# elimina la ultima fila (Totales)
df_avance_por_variedad = df_avance_por_variedad.iloc[:-1]
# elimina las primeras 7 columnas
df_avance_por_variedad = df_avance_por_variedad.iloc[:, 7:].copy()
df_avance_por_variedad.head(5)
```

```python
df_avance_por_variedad = df_avance_por_variedad[df_avance_por_variedad['SIN COSECHA'] > 10]
df_avance_por_variedad
```

```python
df_contactos_grupos.head(3)
```

```python
# JOIN por la columna "id"
df = pd.merge(df_contactos_grupos, df_avance_por_variedad, left_on="COD_COS", right_on="COD_COS", how="inner")
df.head(3)
```

```python
df['fecha_registro'] = datetime.now()
df['cod_canero'] = df['COD_COS']
df['nombre_canero'] = df['NOMBRE']
df['numero_cel'] = df['NUM']

df['mensaje'] = df.apply(lambda row: f"""👉 *Estimado productor, le escribimos desde UTEA por lo siguiente:*

Según el 📡*Monitoreo Satelital* del avance de cosecha 2025 se observa que aún tiene *{row['SIN COSECHA']:.2f} hectáreas de caña de la variedad CITTCA 8522* en campo pendientes para cosechar.

🌱 Le recomendamos priorizar la cosecha de esta variedad lo antes posible por ser una variedad de maduracion temprana, dejarla más tiempo en campo involucra porcentaje de encorchamiento elevado y *perdidas de peso de hasta un 15%*.

🙏 Agradecemos la recepción de este mensaje y su compromiso con el buen manejo de su cosecha. 💪 Estamos seguros de que seguir esta recomendación traerá mejores resultados para su producción. 🚜""", axis=1)


df['enviado'] = False
df['fecha_enviado'] = None
df['numero_contac'] = '591' + df['NUM'].astype(str) + '@s.whatsapp.net'
```

```python
df = df[['fecha_registro', 'cod_canero', 'nombre_canero', 'numero_cel', 'mensaje', 'enviado', 'fecha_enviado', 'numero_contac']]
df
```
