```python
import os
import pandas as pd
```

```python
info = pd.read_excel('G:\BISMARK\_TEMP\CANERO_INST.xlsx')
info
```

```python
lista_inst = list(set(info['INST']))
```

```python
ruta_base = r'C:\Users\Usuario\Desktop\PLANOS_INS\\'

for numero in lista_inst:
    # Crear la ruta de la carpeta
    ruta_carpeta = os.path.join(ruta_base, str(numero))
    
    # Crear la carpeta si no existe
    os.makedirs(ruta_carpeta, exist_ok=True)
    
    print(f"Carpeta creada: {ruta_carpeta}")
```

```python
df_files = pd.DataFrame(os.listdir(), columns=['filename'])
df_files
```

```python
df_files['name'] = df_files['filename'].apply(lambda x : x.split('.')[0].split('_')[1])
df_files
```

```python
for index, row in df_files.iterrows():
    cod_prop = row['name']
    filename = row['filename']
    row = info['COD_PROP']==int(cod_prop)
    new = str(info[row].iloc[0,2]) + '_' + str(info[row].iloc[0,3]) + '_' + str(info[row].iloc[0,0]) + '_' + str(info[row].iloc[0,1])
    inst = info[row].iloc[0,4]
    old_name = r'C:\Users\Usuario\Desktop\PLANOS' + "\\" + filename
    new_name = r'C:\Users\Usuario\Desktop\PLANOS_INS' + "\\" + str(inst) + "\\" + new + '.pdf'
    #print(info[row].iloc[0,4])
    os.rename(old_name, new_name)
    print(new_name)
```
