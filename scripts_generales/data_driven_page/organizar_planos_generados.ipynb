```python
df_files = pd.DataFrame(os.listdir(), columns=['filename'])
df_files
```

```python
df_files['name'] = df_files['filename'].apply(lambda x : x.split('.')[0].split('_')[1])
df_files
```

```python
for index, row in df_files.iterrows():
    cod_prop = row['name']
    filename = row['filename']
    row = info['COD_PROP']==int(cod_prop)
    new = str(info[row].iloc[0,2]) + '_' + str(info[row].iloc[0,3]) + '_' + str(info[row].iloc[0,0]) + '_' + str(info[row].iloc[0,1])
    inst = info[row].iloc[0,4]
    old_name = r'D:\BISMARK\TRABAJOS\ESTIMATIVA_GRAL_2022\PLANOS' + "\\" + filename
    new_name = r'D:\BISMARK\TRABAJOS\ESTIMATIVA_GRAL_2022\PLANOS_INST' + "\\" + str(inst) + "\\" + new + '.pdf'
    #print(info[row].iloc[0,4])
    os.rename(old_name, new_name)
    print(new_name)
```
