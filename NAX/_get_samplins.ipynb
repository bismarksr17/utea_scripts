```python
def crear_usuario_nax(token: str, email: str, password: str, first_name: str,
                       last_name: str, phone: str, company_id: int,
                       apps: list = None, grouping_unit_01_values: list = None,
                       role: str = "basic", language: str = "es-ES"):
    """
    Crea un nuevo usuario en NAX con acceso a ciertas apps
    y limita el acceso a lots específicos según grouping_unit_01.
    """
    url = "https://beta-api.naxsolutions.com/api/v1/users"
    headers = {
        "Authorization": f"Bearer {token}",
        "Content-Type": "application/json"
    }

    # Construir el campo limits
    limits = {
        "companies": [company_id]
    }

    if apps:
        limits["apps"] = apps

    if grouping_unit_01_values:
        limits["lots"] = [
            {
                "company_id": company_id,
                "fields": [
                    {
                        "name": "grouping_unit_01",
                        "values": grouping_unit_01_values
                    }
                ]
            }
        ]

    payload = {
        "email": email,
        "password": password,
        "first_name": first_name,
        "last_name": last_name,
        "phone": phone,
        "role": role,
        "language": language,
        "limits": limits
    }

    response = requests.post(url, headers=headers, data=json.dumps(payload))

    if response.status_code == 200:
        print(f"✅ Usuario creado exitosamente: {email}")
        #return response.json()
    else:
        print(f"❌ Error {response.status_code}: {response.text}")
        return None
```

```python
def listar_usuarios_nax(token: str, corporate_id: int):
    url = f"https://beta-api.naxsolutions.com/api/v1/corporates/{corporate_id}/users"
    headers = {
        "Authorization": f"Bearer {token}"
    }

    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        return response.json()  # Devuelve la lista de usuarios
    else:
        print(f"❌ Error {response.status_code}: {response.text}")
        return []
```
